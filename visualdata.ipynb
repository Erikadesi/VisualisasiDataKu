```python
#1: Plot Garis

import matplotlib.pyplot as plt

#Data
x = [-10, 3, 8, 11, 4]
y = [2, 3, -12, 7, 9]

#Membuat plot garis
plt.plot(x, y, color="orange", label="Garis")

#Menampilkan plot
plt.show()
```

```python
#2: Scatter Plot

import matplotlib.pyplot as plt

#Data
x = [-10, 3, 8, 11, 4]
y = [2, 3, -12, 7, 9]

#Membuat scatter plot
plt.scatter(x, y, color="pink", label="Titik")

#Menampilkan plot
plt.show()
```

```python
#3: Bar Plot

import matplotlib.pyplot as plt

#Data
x = ["Mie Ayam", "Nasi Goreng", "Ayam Penyet", "Bubur Ayam", "Gado-Gado"]
y = [12, 7, 6, 8, 4]

#Membuat bar plot
plt.bar(x, y, color="pink", label="Diagram")

#Menampilkan plot
plt.show()
```

```python
#4: Histogram

import matplotlib.pyplot as plt

#Data
data = [4, 6, 1, 3, 9, 5, 1, 4, 5, 9]

#Membuat histogram (tanpa langsung digambar semua warnanya)
counts, bins, patches = plt.hist(data, bins=5)

#Warna bergantian orange dan pink
colors = ["orange", "pink"]
for i, patch in enumerate(patches):
    patch.set_facecolor(colors[i % 2])

#Menampilkan plot
plt.show()
```

```python
#5: Pie Chart

import matplotlib.pyplot as plt

#Data
labels = ["Mie Ayam", "Nasi Goreng", "Ayam Penyet", "Bubur Ayam", "Gado-Gado"]
sizes = [12, 7, 6, 8, 4]

#Membuat pie chart dengan persentase di dalam
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

#Membuat pie chart
plt.pie(sizes, labels=labels)

#Menampilkan plot
plt.show()
```

```python
#Menampilkan Beberapa Diagram dalam Satu Frame

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Data_Regresi.csv")
fig, axes = plt.subplots(2, 2, figsize=(8, 6), sharex=False)

#Membuat 4 boxplot dengan warna berbeda
p1 = sns.boxplot(data=df, x='Karyawan(X2)', 
                 y='Jual(Y)', width=0.6, ax=axes[0, 0], 
                 color="orange")
p2 = sns.boxplot(data=df, x='Promo(X1)',
                 y='Jual(Y)', width=0.6, ax=axes[0, 1], 
                 color="pink")
p3 = sns.boxplot(data=df, x='Karyawan(X2)', 
                 y='Jual(Y)', width=0.6, ax=axes[1, 0], 
                 color="orange")
p4 = sns.boxplot(data=df, x='Karyawan(X2)',  
                 y='Promo(X1)', width=0.6, ax=axes[1, 1], 
                 color="pink")

plot = [p1, p2, p3, p4]

#Judul tiap subplot
label = ["Karyawan vs Penjualan", 
         "Promo vs Penjualan", 
         "Karyawan vs Penjualan", 
         "Karyawan vs Promo"]

for i in range(4):
    plot[i].set(xlabel=None, ylabel=None)
    plot[i].set_title(label[i], fontsize=11)

fig.tight_layout()
plt.show()
```

Scatter Plot dengan Trendline
</div>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "Data_Regresi.csv"
df = pd.read_csv(url, delimiter=",")
df.columns = df.columns.str.strip()

fig, axes = plt.subplots(2, 2, figsize=(16,10))
plt.subplots_adjust(wspace=0.3, hspace=0.3)

def scatter_with_trend(ax, x, y, color, xlabel, ylabel, title):
    #Scatter
    ax.scatter(x, y, color=color, alpha=0.7, edgecolors="black")

    #Trend line (linear regression)
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    ax.plot(x, p(x), color="black", linewidth=2, linestyle="--", label="Trendline")

    #Label dan style
    ax.set_xlabel(xlabel, fontsize=12)
    ax.set_ylabel(ylabel, fontsize=12)
    ax.set_title(title, fontsize=14)
    ax.grid(alpha=0.3, linestyle="--")
    ax.legend()

#Scatter 1: Promo vs Jual
scatter_with_trend(axes[0,0], df["Promo(X1)"], df["Jual(Y)"],
                   "deeppink", "Promo (X1)", "Jual (Y)", "Promo vs Jual")

#Scatter 2: Karyawan vs Jual
scatter_with_trend(axes[0,1], df["Karyawan(X2)"], df["Jual(Y)"],
                   "hotpink", "Karyawan (X2)", "Jual (Y)", "Karyawan vs Jual")

#Scatter 3: Promo vs Karyawan
scatter_with_trend(axes[1,0], df["Promo(X1)"], df["Karyawan(X2)"],
                   "mediumvioletred", "Promo (X1)", "Karyawan (X2)", "Promo vs Karyawan")

#Scatter 4: TIME vs Jual
scatter_with_trend(axes[1,1], df["TIME"], df["Jual(Y)"],
                   "orchid", "TIME", "Jual (Y)", "TIME vs Jual")

plt.suptitle("Scatter Plot dengan Trendline", fontsize=16, color="crimson")
plt.show()
```
